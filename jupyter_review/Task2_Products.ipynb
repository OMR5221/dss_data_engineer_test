```python
# Get country, city, state, product_name, counts(page_views, visits)
# into new dict:
"""
{"url": "", "breakdown": [
        {"url": "", "counts": ["5", "2"], "name": "USB Disk Security"}, 
        {"url": "", "counts": ["4", "1"], "name": "Free USB Guard"}, 
        {"url": "", "counts": ["2", "1"], "name": "download app"}
    ], "counts": ["11", "2"], "name": "khafji (Ash Sharqlyah, Saudi Arabia)"}
"""

loc_data = []

for prds in products_rd:
    country = prds['name']
    for locs in prds['breakdown']:
        city = locs['name']
        for prods in locs['breakdown']:
            loc_dict = {}
            loc_dict['country'] = country
            loc_dict['city'] = city
            if 'name' in prods: 
                # print(prods)
                loc_dict['product_name'] = prods['name']
                loc_dict['page_views'] = prods['counts'][0]
                loc_dict['visits'] = prods['counts'][1]
                # print("\t\t",prod_name, ": ", page_views, visits)
                loc_data.append(loc_dict)

print(loc_data)
```
